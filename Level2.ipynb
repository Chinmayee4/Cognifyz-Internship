# 2:1 Identify the most common combinations of cuisines in the dataset.

```python
cuisine_combinations = df['Cuisines'].value_counts().head(10).reset_index()  # Top 10 combinations
cuisine_combinations
```

# 2:2 Determine if certain cuisine combinations tend to have higher ratings.

```python
cuisine_ratings=df.groupby('Cuisines')['Aggregate rating'].mean().reset_index().sort_values(by='Aggregate rating',ascending=False).head(10)
cuisine_ratings
```

```python
#plotting it on graph
sns.barplot(x=cuisine_ratings['Aggregate rating'],y=cuisine_ratings['Cuisines'],palette='viridis')
plt.title("Top 10 Cuisine Combinations with Higher Average Ratings")
plt.show()
```

# 3:2 Identify any patterns or clusters of restaurants in specific areas.

```python
from sklearn.cluster import KMeans
df.dropna(subset=['Latitude','Longitude'],inplace=True)
X=df[['Latitude','Longitude']]
k=5
kmeans=KMeans(n_clusters=k,random_state=42)
df['cluster']=kmeans.fit_predict(X)
plt.figure(figsize=(10,6))
plt.scatter(df['Longitude'],df['Latitude'],c=df['cluster'],cmap='viridis',s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Cluster of Restaurants')
plt.colorbar(label='Cluster')
plt.grid(True)
plt.show()
```

# Task: Restaurant Chains

# 4:1 Identify if there are any restaurant chains present in the dataset.

```python
#Counting the no.of restaurants present in the dataset
restaurant_counts=df['Restaurant Name'].value_counts()
restaurant_counts
```

```python
#Identifying restaurant  chains in the dataset
restaurant_chains=restaurant_counts[restaurant_counts>1]
restaurant_chains.reset_index()
```

```python
restaurant_chain_data = df.loc[df['Restaurant Name'].isin(restaurant_chains.index)]
restaurant_chain_data
```

# 4:2 Analyze the ratings and popularity of different restaurant chains.

```python
restaurant_chain_avg_rating = restaurant_chain_data.groupby('Restaurant Name')['Aggregate rating'].mean()
chain_total_votes = restaurant_chain_data.groupby('Restaurant Name')['Votes'].sum()
restaurant_chain_info = pd.DataFrame({'Average rating': restaurant_chain_avg_rating, 'Total votes' : chain_total_votes})
print('Ratings and Popularity of Different Restaurant Chains')
restaurant_chain_info = restaurant_chain_info.sort_values('Average rating', ascending=False)
restaurant_chain_info
```

```python
plt.scatter(x=restaurant_chain_info['Total votes'], y=restaurant_chain_info['Average rating'])
plt.title("Popularity vs rating")
plt.xlabel("Total votes")
plt.ylabel("Avg rating")
plt.show()
```
